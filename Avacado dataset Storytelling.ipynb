```python
import pandas as pd
```

# Loading Dataset

```python
Avacado_data = pd.read_csv("C:\\Users\\vishd\\Desktop\\reetu\\Springboard course\\Project 3\\avocado.csv")
```

```python
Avacado_data.head()
```

```python
Avacado_data = Avacado_data.fillna(0)
```

# Rename col name

```python
Avacado_data = Avacado_data.rename(columns = {'Unnamed: 0': 'Index', '4046': 'Type_4046_sold', '4225': 'Type_4225_sold', '4770': 'Type_4770_sold'})
Avacado_data.head()
```

# Changing Date Object to Datetime 

```python
Avacado_data['Date'] = pd.to_datetime(Avacado_data['Date'], errors='coerce')
```

```python
Avacado_data.info()
```

# Extracting year, month and day from date column 

```python
Avacado_data['year'] = Avacado_data['Date'].dt.year

Avacado_data['month'] = Avacado_data['Date'].dt.month

Avacado_data['day'] = Avacado_data['Date'].dt.day
```

```python
Avacado_data.head()
```

# Set Index

```python
Avacado_data = Avacado_data.reset_index(drop = True)
```

# Avacado price trend ( Price Vs year)

```python
import seaborn as sns

sns.set_theme(style="whitegrid")
sns.set(rc={"figure.figsize":(10, 6)})
#tips = sns.load_dataset("Avacado_data")
ax = sns.lineplot(x="year", y="AveragePrice", data=Avacado_data).set(title='AveragePrice Vs Year')
```

Discussion:
The Price of Avacado was slowly decreasing from 1.375 to 1.330 during 2015 and 2016. 
It suddenly increased after 2016 and went on the peak in 2017 due to more stable production and increasing demand and lower than expected supply. From mid-2017 to the beginning of 2018 average monthly prices shrank back due to a higher supply of Avacado.see below link for more details. 

#https://www.producebluebook.com/2019/02/11/avocado-price-growth-slows/#

# Avacado price trend ( Price Vs Month)

```python
import seaborn as sns

sns.set_theme(style="whitegrid")
sns.set(rc={"figure.figsize":(10, 6)})
#tips = sns.load_dataset("Avacado_data")
ax = sns.lineplot(x="month", y="AveragePrice", data=Avacado_data).set(title='Avacado prive Vs month')
```

Discussion: 

We can see that the price of avocados is jumping from June to Sep. It's all because of the off-bloom season of Avocados(flora loca) which has been supplied from Mexico. Because of the off-season, there is a limited supply of avacado which is causing the increase in the price. Once the Avacado (aventajada) starts to bloom in September, we can see the normal volumes and regular  pricing because of the strong volumes of the Avacado harvesting.

# Month vs bag size

Discussion:

There is no specific trend of total bag production. The trend of total bags and small bags are showing similarities both are increasing in Feb, slipped till April then up in Jun and down till Nov and finally increase in Dec. 

Large bags are showing almost similar trends but the production of larger bags are less than small size bag. XLarge doesn't show any trends but the line show production of XLarge doesn't really relate to weather. Production of Xlarge bags is almost zero throughout the year.
